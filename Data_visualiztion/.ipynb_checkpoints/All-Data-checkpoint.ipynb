```python
## Filling Mising DATA WITH 0

data['price for sales(EGP)'].fillna(0,inplace=True)

data['price for purchases(EGP)'].fillna(0,inplace=True)
```

## EDA

```python
## correlation_matrix to see the correlation between columns
correlation_matrix = data.corr().round(1)
sns.set(rc={'figure.figsize':(8,5)})
sns.heatmap(data=correlation_matrix, annot=True);
```

```python
# Group by 'type' and calculate the sum of sales  within each group
sum_by_type = round(data.groupby('type')['Total Sales(EGP)','Total Purchases(EGP)'].sum())
sum_by_type.plot(kind="bar")
```

```python
sum_by_type = round(data.groupby('type')['purchases(kg)','sales(kg)'].sum())
sum_by_type.plot(kind="bar")
```

```python
sum_by_type = round(data.groupby('type')['Average price','Average price for sales'].mean())
sum_by_type.plot(kind="bar")
```

```python
sum_by_type = round(data.groupby('type')['stock'].mean())
sum_by_type.plot(kind="bar")
```

```python
sum_by_type = round(data.groupby('type')['sales(kg)'].sum())
sum_by_type.plot(kind="pie")
```

```python
data['Date'] = pd.to_datetime(data['Date'])

sales_by_date = data.groupby('Date')['Total Purchases(EGP)'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(sales_by_date['Date'], sales_by_date['Total Purchases(EGP)'])
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Purchases(EGP)')
plt.grid(True)
plt.show()
```

```python

data['Date'] = pd.to_datetime(data['Date'])

sales_by_date = data.groupby('Date')['Total Sales(EGP)'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(sales_by_date['Date'], sales_by_date['Total Sales(EGP)'])
plt.title('Sales Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales(EGP)')
plt.grid(True)
plt.show()
```

```python
data['Date'] = pd.to_datetime(data['Date'])

sales_by_date = data.groupby('Date')['stock'].sum().reset_index()


plt.figure(figsize=(10, 6))
plt.plot(sales_by_date['Date'], sales_by_date['stock'])
plt.title('stock Over Time')
plt.xlabel('Date')
plt.ylabel('stock')
plt.grid(True)
plt.show()
```

## Time seriers 

```python
## cahnge from object to datetime 
data['Date']= pd.to_datetime(data['Date'])
data.dtypes
data.set_index(data.Date , inplace=True) # Set the date to be the index
```

```python
time_ser=data['sales(kg)'].resample('W').mean()

time_ser.plot(figsize=(15, 6))

plt.show()
```

```python
time_ser=data['purchases(kg)'].resample('W').mean()

time_ser.plot(figsize=(15, 6))

plt.show()
```

```python
time_ser=data['stock'].resample('W').mean()

time_ser.plot(figsize=(15, 6))

plt.show()
```
